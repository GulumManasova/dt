# Basic plot

## Line Charts
A Line chart is a graph that represents information as a series of data points connected by a straight line. In line charts, each data point or marker is plotted and connected with a line or curve. 

```python
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [2, 3, 5, 7, 11]

# Plotting the line chart
plt.plot(x, y, marker='o', linestyle='-')

# Adding title and labels
plt.title('Example Line Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Displaying the plot
plt.grid(True)  # Adding grid
plt.show()
```

```python
import matplotlib.pyplot as plt

# Sample data for two lines
x = [1, 2, 3, 4, 5]
y1 = [2, 3, 5, 7, 11]
y2 = [1, 4, 6, 8, 10]

# Plotting the first line
plt.plot(x, y1, marker='o', linestyle='-', label='Line 1')

# Plotting the second line
plt.plot(x, y2, marker='x', linestyle='--', label='Line 2')

# Adding title and labels
plt.title('Multiple Line Chart')
plt.xlabel('X-axis Label')
plt.ylabel('Y-axis Label')

# Adding a legend
plt.legend()

# Displaying the plot
plt.grid(True)  # Adding grid
plt.show()

```

## Bar Graphs
 When you have categorical data, you can represent it with a bar graph. A bar graph plots data with the help of bars, which represent value on the y-axis and category on the x-axi

```python
import matplotlib.pyplot as plt

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 32, 67]

# Plotting the bar graph
plt.bar(categories, values)

# Adding title and labels
plt.title('Example Bar Graph')
plt.xlabel('Categories')
plt.ylabel('Values')

# Displaying the plot
plt.grid(axis='y')  # Adding grid on y-axis
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Sample data
categories = ['A', 'B', 'C', 'D', 'E']
values = [23, 45, 56, 32, 67]

# Calculate average
average = np.mean(values)

# Plotting the bar graph
plt.bar(categories, values, label='Values')

# Plotting the average line
plt.axhline(y=average, color='r', linestyle='--', label='Average')

# Adding title and labels
plt.title('Example Bar Graph with Average')
plt.xlabel('Categories')
plt.ylabel('Values')

# Adding legend
plt.legend()

# Displaying the plot
plt.grid(axis='y')  # Adding grid on y-axis
plt.show()

```

## Scatter Plots
Scatter plots are used when we have to plot two or more variables present at different coordinates. The data is scattered all over the graph and is not confined to a range. Two or more variables are plotted in a Scatter Plot, with each variable being represented by a different color. Let's use the Iris dataset to plot a Scatter Plot.

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Load Iris dataset
iris = sns.load_dataset('iris')

# Plotting scatter plot
sns.scatterplot(x='sepal_length', y='sepal_width', hue='species', data=iris)

# Adding title and labels
plt.title('Sepal Length vs Sepal Width')
plt.xlabel('Sepal Length')
plt.ylabel('Sepal Width')

# Displaying the plot
plt.show()

```

```python
In summary, pie charts are effective for:

Illustrating how different categories contribute to the total (showing parts of a whole).
Comparing proportions or relative sizes of categories.
Highlighting percentages to easily grasp the distribution.
Providing a simple and intuitive visualization suitable for general audiences.
However, it's crucial to use pie charts judiciously, avoiding clutter or excessive categories that can hinder interpretation. In such cases, alternative chart types may be more appropriate.



```

## 3D
Creating 3D plots can be very useful for visualizing complex data in three dimensions. One popular library for 3D plotting in Python is mpl_toolkits.mplot3d from Matplotlib. Here's an example of how you can create a simple 3D scatter plot:

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Generating random data
x = np.random.normal(size=500)
y = np.random.normal(size=500)
z = np.random.normal(size=500)

# Creating 3D scatter plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
ax.scatter(x, y, z)

# Adding labels
ax.set_xlabel('X Label')
ax.set_ylabel('Y Label')
ax.set_zlabel('Z Label')

# Adding title
plt.title('3D Scatter Plot')

# Displaying the plot
plt.show()
```

# Advanced Plot:

Box plot:
In data visualization, a box plot, also known as a box-and-whisker plot, is a method for summarizing the distribution of a dataset. It displays the distribution of a continuous variable through five summary statistics: minimum, first quartile (Q1), median (second quartile or Q2), third quartile (Q3), and maximum. These summary statistics are represented graphically as a box with lines (whiskers) extending from it.

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some random data
np.random.seed(10)
data1 = np.random.normal(loc=0, scale=1, size=100)
data2 = np.random.normal(loc=2, scale=1.5, size=100)
data3 = np.random.normal(loc=-2, scale=0.5, size=100)
data = [data1, data2, data3]

# Create a box plot
plt.figure(figsize=(8, 6))
plt.boxplot(data, labels=['Group 1', 'Group 2', 'Group 3'])
plt.title('Box Plot Example')
plt.xlabel('Groups')
plt.ylabel('Values')
plt.grid(True)
plt.show()
```

## 2D heatmap
A 2D heatmap is a graphical representation of data where values are depicted as colors in a two-dimensional grid. It's often used to visualize the distribution or density of data points across two variables. Matplotlib in Python provides a straightforward way to create 2D heatmaps. 

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate random data
np.random.seed(0)
data = np.random.rand(10, 10)  # 10x10 grid of random values

# Create heatmap
plt.figure(figsize=(8, 6))
plt.imshow(data, cmap='viridis', interpolation='nearest')
plt.colorbar(label='Values')
plt.title('2D Heatmap Example')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.show()
```

## Stacked Bar
A stacked bar chart is a type of chart that represents categorical data with rectangular bars, each divided into segments to show the proportion of each category within a group. Matplotlib in Python can be used to create stacked bar charts. 

```python
import matplotlib.pyplot as plt

# Data
categories = ['Category 1', 'Category 2', 'Category 3']
group_names = ['Group A', 'Group B', 'Group C']
values = [[15, 20, 25],  # Values for Group A
          [10, 25, 20],  # Values for Group B
          [20, 10, 15]]  # Values for Group C

# Create stacked bar chart
fig, ax = plt.subplots(figsize=(8, 6))
width = 0.35
ind = range(len(categories))

bars = []
for i, group_name in enumerate(group_names):
    if i == 0:
        bar = plt.bar(ind, values[i], width, label=group_name)
    else:
        bar = plt.bar(ind, values[i], width, bottom=values[i-1], label=group_name)
    bars.append(bar)

# Add labels, title, legend, etc.
plt.xlabel('Categories')
plt.ylabel('Values')
plt.title('Stacked Bar Chart Example')
plt.xticks(ind, categories)
plt.legend()

plt.show()
```

## Gantt Chart
A Gantt chart is a type of bar chart that illustrates a project schedule, showing the start and finish dates of various elements of the project. Each bar represents a task, with the length of the bar indicating the duration of the task and its position

```python
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# Data
tasks = ['Task 1', 'Task 2', 'Task 3']
start_dates = ['2024-05-01', '2024-05-05', '2024-05-10']
end_dates = ['2024-05-10', '2024-05-15', '2024-05-20']

# Convert dates to datetime objects
start_dates = [mdates.datestr2num(date) for date in start_dates]
end_dates = [mdates.datestr2num(date) for date in end_dates]

# Create Gantt chart
fig, ax = plt.subplots(figsize=(10, 6))

for i, task in enumerate(tasks):
    ax.barh(task, end_dates[i]-start_dates[i], left=start_dates[i], height=0.5, align='center', color='skyblue')

# Format x-axis as dates
ax.xaxis_date()
ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m-%d'))

# Add labels, title, grid, etc.
plt.xlabel('Date')
plt.ylabel('Tasks')
plt.title('Gantt Chart Example')
plt.grid(True)

plt.show()
```

## Polar
Creating a polar plot in Matplotlib allows you to visualize data in a circular or polar coordinate system. Polar plots are particularly useful for representing periodic or cyclical data, such as data distributed around a circle. 

```python
import matplotlib.pyplot as plt
import numpy as np

# Generate some data
theta = np.linspace(0, 2*np.pi, 100)  # Angle values from 0 to 2*pi
r = np.abs(np.sin(2*theta))           # Radius values, absolute value of sin(2*theta)

# Create polar plot
plt.figure(figsize=(8, 6))
plt.polar(theta, r, marker='o')  # 'o' marks data points with circles
plt.title('Polar Plot Example')
plt.show()
```
